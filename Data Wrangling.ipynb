 1. Discovering
 (i)  Read  in  the  Crop_Production  data  from  the  csv  file.  Write  code  to  list  all  column
 names  that  have  trailing  or  leading  spaces,  check  for  inconsistent  spellings  or
 capitalization in categorical columns values like 'District'
 Hints:
 ●  df.info/columns
 ●  df[‘col’].unique

```python
import pandas as pd

cp = pd.read_csv("Crop_Production.csv")
cp.info()

dim = cp.shape
print("Dimensions of the crop production dataset: ", dim)
```

```python
cols = cp.columns
print(cols)

cp.columns = [col.strip() for col in cp.columns]

print("Updated Column names: ", cp.columns)
```

```python
cp['State'].unique()
cp['State'] = cp['State'].str.replace(" ","_")
cp['State']
```

```python
cp['District'] = cp['District'].str.capitalize()
cp['District']
```

(ii)  Find the top 5 highest 'Yield' crops on average across the entire dataset.
 Hints  :
 ●  df.groupby:Splitting, Applying, and Combining.
 ●  mean(),  sort_values(), head()

```python
top_5 = cp.groupby('Crop')['Yield'].mean().sort_values(ascending=False).head(5)
top_5
```

(iii)  Find all crop names ('Crop') that appear fewer than 300 times in the entire dataset.
 Hints  :
 ●  value_count()

```python
count = cp['Crop'].value_counts()
final_count = count[count<300]
print(final_count)
```

 2.Structuring
 (i)  Rename all column names that have trailing or leading spaces
 Hints  :
 ●  df.rename(columns={oldname,newname})

```python
cp.rename(columns={"District ": "District"}, inplace=True)
```

 (ii)  What  is  the  seasonal  performance  of  each  crop,  in  terms  of  both  total  Production  and
 average Yield
 Hints  :
 ●  df.groupby a list of column names
 ●  agg():   apply one or more aggregation functions to one or more columns,

```python
seasons = cp.groupby(['Crop','Season']).agg(
    total_production=('Production', 'sum'),
    average_yield=('Yield', 'mean')
)
seasons
```

 (iii)  What is the year-wise total production for each crop
 Hints  :
 ●  df.pivot(): transform a DataFrame from a "long" format to a "wide" format
 Arguments: index, columns, and values.

```python
year_wise_production_pivot = pd.pivot_table(
    cp,
    index='Crop',
    columns='Crop_Year',
    values='Production',
    aggfunc='sum'
)
year_wise_production_pivot
```

 (iv)  Which crop, in which year, produced the highest average yield
 Hints  :
 ●  df.melt(): transform or reshape DataFrames from a "wide" format to a "long"
 format.
 Arguments: id_vars, value_vars, var_name, value_name
 [use transformed data from pivot)

```python
final = pd.melt(year_wise_production_pivot.reset_index(), id_vars=['Crop'], value_vars=[2010,1997,1998], var_name='Production')
final
```

3. Cleaning
(i) Change inconsistent columns values like 'State' and 'District' from uppercase to
lowercase and remove trailing or leading spaces
Hints:
● df[‘col’].str.lower(), df[‘col’].str.strip()

```python
cp['State'] = cp['State'].str.lower()
cp['State'] = cp['State'].str.strip()
cp['District'] = cp['District'].str.lower()
cp['District'] = cp['District'].str.strip()
```

(ii) Instead of filling missing Production with 0 or a global mean, impute the NaN values
by calculating Area * (Average_Yield). The Average_Yield should be the average for
that specific District and Crop combination.
Hints:
● groupby(['District', 'Crop']), transform('mean'), and fillna()).

```python
cp['group'] = cp.groupby(['District', 'Crop'])['Yield'].transform('mean')
imputed = cp['Area'] * cp['group']
cp['Production'] = cp['Production'].fillna(imputed)
```

(iii) Find the wheat production of Odisha state
Hints:
● filter by Multiple Conditions (AND - &)

```python
odisha = cp[(cp['Crop'] == "Wheat") & (cp['State'] == "odisha")]
odisha
```

4. Enrichment
(i) Merge with External Data
Hints:
● df.merge(data1,data2)
Check whether common column and its values are compatible or not before
merge.

```python
table1 = pd.read_csv("Crop_Production1.csv")
table2 = pd.read_csv("city_day.csv")

table1.rename(columns = {"District": "city"}, inplace=True)
table2.rename(columns = {"City": "city"}, inplace=True)

all_names= set(table1['city']) | set(table2['city'])

dict = {
    'KOLKATA': 'Kolkata',
    'SHILLONG': 'Shillong',
    'MUMBAI': 'Mumbai',
    'JAIPUR': 'Jaipur',
    'AIZAWL': 'Aizawl',
    'KOCHI': 'Kochi',
    'THIRUVANANTHAPURAM': 'Thiruvananthapuram',
    'BENGALURU': 'Bengaluru',
    'GURUGRAM': 'Gurugram',
    'AHMADABAD': 'Ahmedabad',
    'AHMEDABAD': 'Ahmedabad',
    'DELHI_TOTAL': 'Delhi',
    'DELHI': 'Delhi',
    'AMRITSAR': 'Amritsar',
    'PATNA': 'Patna',
    'CHENNAI': 'Chennai',
    'LUCKNOW': 'Lucknow',
    'HYDERABAD': 'Hyderabad',
    'COIMBATORE': 'Coimbatore',
    'BHOPAL': 'Bhopal',
    'VISAKHAPATNAM': 'Visakhapatnam',
    'BRAJRAJNAGAR': 'Brajrajnagar',
    'ERNAKULAM': 'Ernakulam',
    'BANGALORERURAL': 'Bangalore Rural',
    'AMARAVATI': 'Amaravati',
    'GUWAHATI': 'Guwahati',
    'CHANDIGARH': 'Chandigarh',
    'TALCHER': 'Talcher',
    'JORAPOKHAR': 'Jorapokhar'
}

table1['city'] = table1['city'].replace(dict)
table1['city'] = table1['city'].str.lower()
table1['city'] = table1['city'].str.strip()

table2['city'] = table2['city'].replace(dict)
table2['city'] = table2['city'].str.lower()
table2['city'] = table2['city'].str.strip()

table3 = pd.merge(table1, table2, on='city')
table3
```

5. Validating
(i) The Yield column should equal Production / Area. Create a new column called
Calculated_Yield using this formula. Then, find the sum of the difference between
Calculated_Yield and Yield to see if the original data was accurate.

```python
table3['Cal_Yield'] = table3['Production']/table3['Area']
table3['Difference'] = table3['Cal_Yield'] - table3['Yield']
sum = table3['Difference'].sum()
sum
```

(ii) Check if any Area, Production, or Yield values are negative.
Check if Production > 0 and Area == 0
Check if any Crop_Year is in the future (e.g., > 2024).

```python
isNeg = table3[(table3['Area'] < 0) | (table3['Production'] < 0) | (table3['Yield'] < 0)]
if isNeg.empty:
    print("No Negative values found.")
else:
    print("Negative values found.")

zero = table3[(table3['Production'] > 0) & (table3['Area'] == 0)]
if zero.empty:
    print("\nRows with Production > 0 and Area == 0 not found.")
else:
    print("\nRows with Production > 0 and Area == 0 found.")

futureCropYear = table3[table3['Crop_Year'] > 2024]
if futureCropYear.empty:
    print("\nRows with Crop_Year in the future (> 2024) not found.")
else:
    print("\nRows with Crop_Year in the future (> 2024) found.")
```
